### Paging

```python
paging = result_soup.find('div',{'id':'pagination'})
```

```python
pages = paging.find_all('a')
len(pages)
```

```python
next_page = None
for page in pages:
    if page.text.strip()=='∨':
        next_page = page.get('href')
        break
```

```python
start_urls = ['http://bundes-telefonbuch.de/']
if next_page:
    next_page = start_urls[0] + next_page
print(next_page)
```
